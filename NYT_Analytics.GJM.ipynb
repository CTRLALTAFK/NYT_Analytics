```python
# Dependencies
import csv
import matplotlib.pyplot as plt
import requests
from pprint import pprint
import pandas as pd
import json
from config import api_key
#url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?q=&api-key="
#api_key = "../API_Keys"
base_url = "https://api.nytimes.com/svc/mostpopular/v2/"
emailed = "emailed/"
shared = "shared/"
viewed = "viewed/"
time_1 = "1.json?"
time_7 = "7.json?"
time_30 = "30.json?"

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"
begin_date = "20190101"
end_date = "20190930"
query_url4a = f"{url}api-key={api_key}&begin_date={begin_date}&end_date={end_date}"
baseurl4a = query_url4a
```

```python
############################################ Most EMAILED - ##############################################


# # baseurl1 = emailed = https://api.nytimes.com/svc/mostpopular/v2/emailed/7.json?api-key=yourkey
# # Retrieving data and converting it into JSON
#pprint(requests.get(new_url1).json())

# Most EMAILED - access

query_url = base_url + emailed + time_30 + "api-key=" + api_key

articles = requests.get(query_url).json()
articles = articles['results']

pprint(articles)
```

```python
# Most emailed cont.

#prepare lists
title = []
section = []
#keywords = []
pop_rank = []

#fill lists with data
for article in articles:
    title.append(article['title'])
    section.append(article['section'])
    #keywords.append(article['des_facet'])
    pop_rank.append(article['email_count'])

#convert lists to dict & dataframe
emailed_30 = {'Title': title, 'Section': section, 'Popularity Rank': pop_rank}
emailed_30 = pd.DataFrame(emailed_30)
emailed_30
```

```python
# Most emailed - Graph it
section_pop1 = emailed_30['Section'].value_counts()
section_pop1a=section_pop1.plot(kind = "bar")

# Set the xlabel and ylabel using class methods
section_pop1a.set_xlabel("Genre of Email")
section_pop1a.set_ylabel("Times Emailed")
```
